# Metodologa de muestreo: El RNPDNO y la crisis de desapariciones

Para este anlisis, utilizamos los datos del **Registro Nacional de Personas Desaparecidas y No Localizadas (RNPDNO)** con corte a enero de 2026. La estadstica nos permite extraer una muestra representativa para estudiar patrones de comportamiento y sesgos de datos en los registros oficiales.

**Parmetros de muestreo:**
* **Poblacin ($N$):** 132,212 personas.
* **Nivel de Confianza ($z$):** 95% (1.96).
* **Probabilidad de xito ($p$):** 0.5.
* **Margen de Error ($e$):** 5% (0.05).

```python
# Definicin de variables
N = 132212
z = 1.96
p = 0.5
q = 1 - p
e = 0.05

# 1. Clculo del tamao de muestra preliminar (poblacin infinita)
n0 = (z**2 * p * q) / (e**2)

# 2. Aplicacin de la frmula de correccin por poblacin finita
n = (N * n0) / (n0 + N - 1)

print(f"Tamao de muestra preliminar (n0): {n0:.2f}")
print(f"Tamao de muestra corregido (n): {n:.2f}")
print(f"Muestra final representativa: {round(n)} personas")
```

## Desarrollo del clculo

Para determinar el tamao de la muestra, primero calculamos el valor base para una poblacin infinita mediante la frmula:

$$n_0 = \frac{z^2 \cdot p \cdot (1-p)}{e^2}$$

Sustituyendo nuestros valores:
$$n_0 = \frac{1.96^2 \cdot 0.5 \cdot 0.5}{0.0025} = 384.16$$

Posteriormente, aplicamos la **correccin por poblacin finita** para ajustar el tamao de la muestra al universo real de casos en Mxico ($N = 132,212$):

$$n = \frac{N \cdot n_0}{n_0 + N - 1}$$

$$n = \frac{132,212 \cdot 384.16}{384.16 + 132,212 - 1} \approx 383.05$$

**Resultado:** Para que el estudio sea estadsticamente representativo, se requiere una muestra de **383 personas**.

## Anlisis desde la perspectiva de data fairness

Aunque 383 casos otorgan validez estadstica a nivel nacional, desde la ptica del **AI Engineering** y el **Data Fairness**, debemos cuestionar:

1. **Sesgos regionales:** Es esta muestra suficiente para representar estados con crisis agudas comparados con estados con bajo reporte?
2. **Invisibilidad de datos:** El valor de $N$ representa solo los casos reportados. Un modelo entrenado solo con esta muestra podra heredar el sesgo de las autoridades que recolectan la informacin.
3. **Representatividad especfica:** En modelos de deteccin de sesgos, a veces es necesario un sobremuestreo (*oversampling*) de grupos minoritarios para asegurar que el algoritmo no sea injusto con ellos.
